```python
# line plot data - set 1
list_x = [1, 2, 3]
list_y = [2, 4, 6]

# line plot data - set 2
list_x2 = [1, 2, 3]
list_y2 = [7, 7, 14]

# plot two sets of data
plt.plot(list_x, list_y, label = "First Line")
plt.plot(list_x2, list_y2, label = "Second Line")

# show data
plt.show()
```

```python
# line plot data - set 1
list_x = [1, 2, 3]
list_y = [2, 4, 6]

# line plot data - set 2
list_x2 = [1, 2, 3]
list_y2 = [7, 7, 14]

# plot two sets of data
plt.plot(list_x, list_y, label = "First Line")
plt.plot(list_x2, list_y2, label = "Second Line")

# add title and axis labels
plt.title("My Plot")
plt.xlabel("My X label")
plt.ylabel("My Y label")

# add legend
plt.legend(loc = "center right")

# show data
plt.show()
```
